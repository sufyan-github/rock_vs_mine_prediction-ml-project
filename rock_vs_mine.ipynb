Import the Dependencies

```python
#import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

print("Import Libraries and Dependencies Successfully")
```

Data Collection and Data Processing

```python
#Dataset
sonar_data = pd.read_csv('sonar data.csv', header=None)
sonar_data.head()
```

```python
sonar_data.shape
```

```python
sonar_data.describe()
```

```python
sonar_data[60].value_counts()
```

```python
sonar_data.groupby(60).mean()
```

Data separating with features and target

```python
X = sonar_data.drop(columns=60, axis=1)
y = sonar_data[60]
```

Training and test data split

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, stratify=y, random_state=1)

print(X.shape, X_train.shape, X_test.shape)
```

Model Training

```python
model = LogisticRegression()
model.fit(X_train, y_train)
```

Model Evaluation

```python
#Accuracy on training data
score = model.score(X_train, y_train)
print("Training Accuracy:", score)
```

```python
x_training_prediction = model.predict(X_train)
accuracy = accuracy_score(y_train, x_training_prediction)
print("Training Accuracy:", accuracy) 
```

Test Prediction

```python
#accuracy on test data
x_test_prediction = model.predict(X_test)
test_accuracy = accuracy_score(y_test, x_test_prediction)
print("Test Accuracy:", test_accuracy)
```

```python
#plotting prediction and actual values

plt.figure(figsize=(12, 6))

# Plotting training data predictions
plt.subplot(1, 2, 1)
plt.scatter(y_train, x_training_prediction)
plt.title("Training Data: Actual vs Predicted")
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.plot([0, 1], [0, 1], 'r--')  # Diagonal line

# Plotting test data predictions
plt.subplot(1, 2, 2)
plt.scatter(y_test, x_test_prediction)
plt.title("Test Data: Actual vs Predicted")
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.plot([0, 1], [0, 1], 'r--')  # Diagonal line

plt.tight_layout()
plt.show()
```
